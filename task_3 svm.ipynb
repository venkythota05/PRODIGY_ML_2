```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Model
from tensorflow.keras.applications.vgg16 import preprocess_input
import random
```

```python
base_model = VGG16(weights="imagenet", include_top=False, input_shape=(128, 128, 3))
model = Model(inputs=base_model.input, outputs=base_model.output)
```

```python
def load_data(directory, label, img_size=(128, 128)):
    images = []
    labels = []
    filenames = []

    for filename in os.listdir(directory):
        img_path = os.path.join(directory, filename)
        img = cv2.imread(img_path)
        if img is not None:
            img = cv2.resize(img, img_size)
            img = img / 255.0  # Normalize pixel values
            images.append(img)
            labels.append(label)
            filenames.append(filename)  # Changed line: Add filename to the list
    return images, labels, filenames
```

```python
cats, labels_cats, filenames_cats = load_data("/content/drive/MyDrive/dogvscat/train/ctrain", 0)
dogs, labels_dogs, filenames_dogs = load_data("/content/drive/MyDrive/dogvscat/train/dtrain", 1)
```

```python
X = np.array(cats + dogs)
y = np.array(labels_cats + labels_dogs)
filenames = np.array(filenames_cats + filenames_dogs)
```

```python
X_features = model.predict(X)
X_features = X_features.reshape(X_features.shape[0], -1)
```

```python
X_train, X_test, y_train, y_test, filenames_train, filenames_test = train_test_split(
    X_features, y, filenames, test_size=0.2, random_state=42
)
```

```python
svm = SVC(kernel="linear")
svm.fit(X_train, y_train)
```

```python
y_pred = svm.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=["Cat", "Dog"]))
```
